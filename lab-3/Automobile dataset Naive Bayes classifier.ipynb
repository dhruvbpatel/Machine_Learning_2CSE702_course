```python
## splitting dataset into feature and target columns
```

```python
y = df_numeric['symboling']   # target columns
```

```python
y
```

```python
df_numeric.drop(['symboling'],axis=1,inplace=True)  
```

```python
X = df_numeric  # feature columns
```

```python
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state = 3)  #splitting dataset
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
## buidling model
```

```python
df['normalized-losses'].value_counts()
```

```python
gnb = GaussianNB()   # initialize model
```

```python
gnb.fit(X_train,y_train)   ## fit model on train data
```

```python
y_pred = gnb.predict(X_test)  # get y_pred
```

```python
print(accuracy_score(y_test,y_pred))   # accuracy score
```

```python
print(confusion_matrix(y_test,y_pred))  ## our multiclass confusion metrics
```

```python
y_pred  ## y_pred
```

```python
print(recall_score(y_test,y_pred,average=None)) # recall score
```

```python
print(precision_score(y_test,y_pred,average=None))  # precision score
```

```python
print(f1_score(y_test,y_pred,average=None))  # f1_score
```
